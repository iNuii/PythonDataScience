# **Function**

```python
def addonetonum(num):
  output = num + 1
  return output

addonetonum(9)
```

```python
def sumtwonumbers(num1, num2):
  output = num1 + num2
  return output

sumtwonumbers(15, 17)
```

```python
def getsumfromlist(numList):
  totalsum = 0
  for num in numList:
    totalsum += num
  return totalsum
```

```python
numList = [1, 2, 3, 4, 5, 6]
getsumfromlist(numList)
```

```python
def getmeanfromlist(numList):
  totalsum = 0
  totalnum = 0
  for num in numList:
    totalsum += num
    totalnum += 1
  meanlist = totalsum/totalnum
  return meanlist
```

```python
numList = [3, 4, 3, 2, 4, 4, 3]
getmeanfromlist(numList)
```

# Game Assesment

```python
def guessMyNumber():
  myNum = 10
  output = ''
  while output != 'correct!':
    print('Please input your guese.')
    numguese = int(input())
    if numguese == myNum:
      output = 'correct!'
    else:
      if numguese < myNum:
        output = 'Your guess is smaller.'
      else:
        output = 'Your guess is larger.'
    print(output)
    print('\n')
  return output
```

```python
guessMyNumber()
```
